```python
# Data Quality Assessment & Profiling 2

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (update path or URL as needed)
df = pd.read_csv('your_dataset.csv')  # Example: 'data/data2.csv'

# Preview the dataset
print("First 5 rows:")
print(df.head())

# Dataset shape and basic info
print(f"\nDataset Shape: {df.shape}")
print("\nColumn Names:")
print(df.columns.tolist())
print("\nData Types:")
print(df.dtypes)

# Summary statistics
print("\nSummary Statistics:")
print(df.describe(include='all'))

# Missing values check
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_report = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
})
print("\nMissing Value Report:")
print(missing_report)

# Duplicate row check
duplicate_rows = df.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicate_rows}")

# Unique values and inconsistency detection
print("\nUnique Values per Column (showing up to 10):")
for col in df.columns:
    print(f"\n{col}:")
    print(df[col].unique()[:10])  # First 10 unique values

# Value counts for categorical variables
print("\nCategorical Columns Value Counts:")
categorical_cols = df.select_dtypes(include='object')
for col in categorical_cols.columns:
    print(f"\n{col} Value Counts:")
    print(df[col].value_counts())

# Outlier detection using boxplots
print("\nBoxplots for Numerical Columns:")
numeric_cols = df.select_dtypes(include=np.number)
for col in numeric_cols.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()

# Final Data Quality Report
quality_report = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Missing Values': df.isnull().sum().values,
    'Unique Values': df.nunique().values,
    'Duplicate Count': [df[col].duplicated().sum() for col in df.columns]
})
print("\nData Quality Report:")
print(quality_report)
```
