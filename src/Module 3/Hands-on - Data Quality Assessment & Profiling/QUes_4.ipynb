```python
# Data Quality Assessment & Profiling 4

# Step 1: Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv('your_dataset.csv')  # Replace with actual path or filename

# Step 3: View first few rows
print("First 5 rows of the dataset:")
print(df.head())

# Step 4: Dataset structure
print(f"\nDataset Shape: {df.shape}")
print("\nColumn Names:")
print(df.columns.tolist())
print("\nData Types:")
print(df.dtypes)

# Step 5: Summary statistics
print("\nSummary Statistics (numeric + categorical):")
print(df.describe(include='all'))

# Step 6: Missing value analysis
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_report = pd.DataFrame({
    'Missing Values': missing_values,
    'Missing Percentage (%)': missing_percent
})
print("\nMissing Value Report:")
print(missing_report)

# Step 7: Duplicate rows check
duplicate_count = df.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicate_count}")

# Step 8: Unique value analysis
print("\nUnique Values per Column (top 10 shown):")
for col in df.columns:
    print(f"\n{col}:")
    print(df[col].unique()[:10])

# Step 9: Categorical columns value counts
cat_cols = df.select_dtypes(include='object')
print("\nCategorical Columns Frequency:")
for col in cat_cols.columns:
    print(f"\n{col}:\n{df[col].value_counts()}")

# Step 10: Outlier detection using boxplots
print("\nGenerating Boxplots for Numerical Columns:")
num_cols = df.select_dtypes(include=np.number)
for col in num_cols.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot - {col}')
    plt.show()

# Step 11: Final data quality summary report
quality_report = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Missing Values': df.isnull().sum().values,
    'Unique Values': df.nunique().values,
    'Duplicate Column-wise': [df[col].duplicated().sum() for col in df.columns]
})
print("\nFinal Data Quality Report:")
print(quality_report)
```
