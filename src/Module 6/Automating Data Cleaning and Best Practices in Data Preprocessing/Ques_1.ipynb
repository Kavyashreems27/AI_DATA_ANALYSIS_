```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np

# Step 2: Define data cleaning functions
def standardize_column_names(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    return df

def handle_missing_values(df):
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col].fillna(df[col].mean(), inplace=True)
    for col in df.select_dtypes(include='object').columns:
        df[col].fillna(df[col].mode()[0], inplace=True)
    return df

def remove_duplicates(df):
    return df.drop_duplicates()

def remove_outliers(df):
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

def convert_data_types(df):
    # Example: Convert a column to datetime if applicable
    # df['date_column'] = pd.to_datetime(df['date_column'])
    return df

# Step 3: Create the pandas pipeline
def data_cleaning_pipeline(df):
    return (
        df.pipe(standardize_column_names)
          .pipe(handle_missing_values)
          .pipe(remove_duplicates)
          .pipe(remove_outliers)
          .pipe(convert_data_types)
    )

# Step 4: Load dataset and apply pipeline
# Replace 'your_dataset.csv' with actual file path
df = pd.read_csv('your_dataset.csv')
df_cleaned = data_cleaning_pipeline(df)

# Step 5: Display cleaned data
print("Cleaned Data:")
print(df_cleaned.head())

# Step 6: Save to new CSV (Optional)
df_cleaned.to_csv('cleaned_data.csv', index=False)
```
