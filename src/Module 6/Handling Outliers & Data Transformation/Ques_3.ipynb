```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats.mstats import winsorize

# Step 1: Generate sample data with outliers
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Generate 100 data points
outliers = np.array([100, 105, 110])  # Define outliers
data_with_outliers = np.concatenate((data, outliers))  # Combine data

# Convert to a pandas DataFrame
df = pd.DataFrame(data_with_outliers, columns=['Value'])

# Step 2: Visualize the data before Winsorization
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Value'])
plt.title('Boxplot Before Winsorization')
plt.ylabel('Value')
plt.show()

# Step 3: Apply Winsorization
# Winsorize the data by capping the lowest and highest 5% of the data
winsorized_data = winsorize(df['Value'], limits=[0.05, 0.05])

# Add the Winsorized data to the DataFrame
df['Winsorized_Value'] = winsorized_data

# Step 4: Visualize the data after Winsorization
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Winsorized_Value'])
plt.title('Boxplot After Winsorization')
plt.ylabel('Winsorized Value')
plt.show()

# Step 5: Compare statistical summaries before and after Winsorization
print("Statistical Summary Before Winsorization:")
print(df['Value'].describe())

print("\nStatistical Summary After Winsorization:")
print(df['Winsorized_Value'].describe())
```
