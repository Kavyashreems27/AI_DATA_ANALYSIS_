```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample

# Step 1: Load the Credit Card Fraud Detection dataset
# Dataset URL (Kaggle): https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
# For this example, we assume the file is named 'creditcard.csv' and in the same directory
df = pd.read_csv('creditcard.csv')

# Step 2: Visualize original class distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Class')
plt.title('Original Class Distribution')
plt.xlabel('Class (0 = Legit, 1 = Fraud)')
plt.ylabel('Count')
plt.show()

print("Original class counts:")
print(df['Class'].value_counts())

# Step 3: Handle imbalance using random undersampling
# Separate majority and minority classes
df_majority = df[df['Class'] == 0]
df_minority = df[df['Class'] == 1]

# Undersample majority class
df_majority_undersampled = resample(df_majority,
                                    replace=False,
                                    n_samples=len(df_minority),
                                    random_state=42)

# Combine minority class with undersampled majority class
df_balanced = pd.concat([df_majority_undersampled, df_minority])

# Step 4: Visualize balanced class distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df_balanced, x='Class')
plt.title('Balanced Class Distribution (After Undersampling)')
plt.xlabel('Class (0 = Legit, 1 = Fraud)')
plt.ylabel('Count')
plt.show()

print("Balanced class counts:")
print(df_balanced['Class'].value_counts())
```
