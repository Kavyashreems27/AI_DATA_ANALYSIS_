```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

# Step 1: Simulate Dataset A (original)
np.random.seed(0)
data_a = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Simulate Dataset B (shifted)
data_b = np.random.normal(loc=0.5, scale=1, size=1000)  # shift in mean

# Step 3: Plot both distributions
plt.hist(data_a, bins=30, alpha=0.5, label='Dataset A (Original)', density=True)
plt.hist(data_b, bins=30, alpha=0.5, label='Dataset B (Shifted)', density=True)
plt.legend()
plt.title("Distribution Comparison: Data A vs Data B")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Step 4: Apply Kolmogorovâ€“Smirnov test
ks_stat, p_value = ks_2samp(data_a, data_b)

# Step 5: Print result
print(f"KS Statistic: {ks_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
if p_value < 0.05:
    print("Result: Data drift detected (distributions are different).")
else:
    print("Result: No significant data drift detected.")
```
