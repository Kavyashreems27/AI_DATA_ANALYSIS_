```python
import pandas as pd
import seaborn as sns

# Step 1: Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')

# Step 2: Display initial info and count of missing values
print("Initial Dataset Info:")
print(df.info())
print("\nMissing Values Per Column:")
print(df.isnull().sum())

# Step 3: Handle missing values

# Example strategies:
# - Fill 'age' with median
# - Fill 'embarked' with mode
# - Drop 'deck' due to too many missing values
# - Drop rows where 'embark_town' is missing

df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df.drop(columns=['deck'], inplace=True)
df.dropna(subset=['embark_town'], inplace=True)

# Step 4: Confirm that missing values are handled
print("\nAfter Cleaning Missing Values:")
print(df.isnull().sum())
```
