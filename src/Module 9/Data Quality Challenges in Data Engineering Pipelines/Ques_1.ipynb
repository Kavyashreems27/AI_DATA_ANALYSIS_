```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {
    'Age': [25, np.nan, 30, np.nan, 29, 27, 32, 31],
    'Fare': [100, 200, np.nan, 150, 120, np.nan, 180, 160],
    'Gender': ['Male', 'Female', np.nan, 'Female', 'Male', 'Male', np.nan, 'Female']
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame with Missing Values:")
print(df)

# Impute missing values in 'Age' column with mean
df['Age_mean_imputed'] = df['Age'].fillna(df['Age'].mean())

# Impute missing values in 'Fare' column with median
df['Fare_median_imputed'] = df['Fare'].fillna(df['Fare'].median())

# Impute missing values in 'Gender' column with mode
df['Gender_mode_imputed'] = df['Gender'].fillna(df['Gender'].mode()[0])

# Display the DataFrame after imputation
print("\nDataFrame after Imputation:")
print(df)
```
