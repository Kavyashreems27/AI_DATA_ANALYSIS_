```python
import numpy as np
import pandas as pd
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Simulate two datasets with a feature and drift
np.random.seed(42)

# Original distribution (reference data)
data_ref = pd.DataFrame({
    'feature': np.random.normal(loc=0, scale=1, size=1000)
})

# Drifted distribution (new data)
data_new = pd.DataFrame({
    'feature': np.random.normal(loc=0.5, scale=1.2, size=1000)
})

# Step 2: Apply Kolmogorov-Smirnov test
statistic, p_value = ks_2samp(data_ref['feature'], data_new['feature'])

# Step 3: Output the result
print("Kolmogorov-Smirnov Test Results:")
print(f"Statistic: {statistic:.4f}")
print(f"P-Value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion: Significant data drift detected.")
else:
    print("Conclusion: No significant data drift detected.")

# Step 4: Visualize the distributions
plt.figure(figsize=(10, 6))
sns.kdeplot(data_ref['feature'], label='Reference Data', fill=True)
sns.kdeplot(data_new['feature'], label='New Data', fill=True)
plt.title('Feature Distribution Comparison')
plt.xlabel('Feature Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
