```python
import pandas as pd
import re
from datetime import datetime

# Load your dataset
# Replace 'your_data.csv' with the path to your actual dataset
df = pd.read_csv("your_data.csv")

# 1. Uniqueness Check: Identify duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
if not duplicate_rows.empty:
    print("Duplicate Rows:")
    print(duplicate_rows)

# 2. Uniqueness Check: Check for unique values in specific columns
# Replace 'ColumnName' with the actual column name you want to check
unique_counts = df['ColumnName'].nunique()
total_counts = df['ColumnName'].count()
print(f"Unique values in 'ColumnName': {unique_counts} out of {total_counts}")

# 3. Validity Check: Email Format Validation
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
df['Email_Valid'] = df['Email'].apply(lambda x: bool(re.match(email_pattern, str(x))))
invalid_emails = df[~df['Email_Valid']]
print(f"Number of invalid emails: {invalid_emails.shape[0]}")
if not invalid_emails.empty:
    print("Invalid Emails:")
    print(invalid_emails[['Email']])

# 4. Validity Check: Date Format Validation (YYYY-MM-DD)
def is_valid_date(date_str):
    try:
        datetime.strptime(str(date_str), '%Y-%m-%d')
        return True
    except:
        return False

df['Date_Valid'] = df['Date'].apply(is_valid_date)
invalid_dates = df[~df['Date_Valid']]
print(f"Number of invalid dates: {invalid_dates.shape[0]}")
if not invalid_dates.empty:
    print("Invalid Dates:")
    print(invalid_dates[['Date']])

# 5. Validity Check: Phone Number Validation (10-digit numbers)
phone_pattern = r'^\d{10}$'
df['Phone_Valid'] = df['Phone'].apply(lambda x: bool(re.match(phone_pattern, str(x))))
invalid_phones = df[~df['Phone_Valid']]
print(f"Number of invalid phone numbers: {invalid_phones.shape[0]}")
if not invalid_phones.empty:
    print("Invalid Phone Numbers:")
    print(invalid_phones[['Phone']])

# Optional: Save the validated dataset to a new CSV file
df.to_csv("validated_output.csv", index=False)
print("\nData quality assessment completed. Results saved to 'validated_output.csv'.")
```
