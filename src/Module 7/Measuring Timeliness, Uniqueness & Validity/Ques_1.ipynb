```python
import pandas as pd
import re
from datetime import datetime

# Load the dataset (update the file name as needed)
# Example: df = pd.read_csv("data.csv")
df = pd.read_csv("your_data.csv")  # Replace with your actual file

# --- FORMAT VALIDATION FUNCTIONS ---

# 1. Email validation
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
df['Email_Valid'] = df['Email'].apply(lambda x: bool(re.match(email_pattern, str(x))))

# 2. Date validation (format: YYYY-MM-DD)
def is_valid_date(date_str):
    try:
        datetime.strptime(str(date_str), '%Y-%m-%d')
        return True
    except:
        return False
df['Date_Valid'] = df['Date'].apply(is_valid_date)

# 3. Phone number validation (10-digit numbers)
phone_pattern = r'^\d{10}$'
df['Phone_Valid'] = df['Phone'].apply(lambda x: bool(re.match(phone_pattern, str(x))))

# --- OUTPUT RESULTS ---

# View invalid entries
invalid_emails = df[~df['Email_Valid']]
invalid_dates = df[~df['Date_Valid']]
invalid_phones = df[~df['Phone_Valid']]

print("Invalid Emails:\n", invalid_emails[['Email']])
print("\nInvalid Dates:\n", invalid_dates[['Date']])
print("\nInvalid Phone Numbers:\n", invalid_phones[['Phone']])

# Optional: Save to CSV for reference
df.to_csv("validated_output.csv", index=False)
```
