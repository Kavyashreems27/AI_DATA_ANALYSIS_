```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Sample sales data
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [100, 150, 90, 120]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bar plot
plt.figure(figsize=(8,6))
plt.bar(df['Product'], df['Sales'], color='orange')  # Changed color to orange

# Customize fonts
plt.title('Sales Comparison', fontsize=16, fontweight='bold', fontfamily='serif')
plt.xlabel('Product', fontsize=14, fontfamily='serif')
plt.ylabel('Sales', fontsize=14, fontfamily='serif')

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
