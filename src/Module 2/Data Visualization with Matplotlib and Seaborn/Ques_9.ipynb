```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Data for the bar plot (Product Sales)
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [120, 180, 95, 210]

# Data for the scatter plot (Height vs. Weight)
height = [165, 170, 155, 180, 160]
weight = [60, 75, 50, 85, 55]
size_data = [10, 15, 13, 18, 12]
data_scatter = pd.DataFrame({'Height': height, 'Weight': weight, 'Size': size_data})

# Create the figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6)) # 1 row, 2 columns of subplots

# Bar plot (first subplot)
axes[0].bar(products, sales, color='skyblue')
axes[0].set_title('Product Sales')
axes[0].set_xlabel('Products')
axes[0].set_ylabel('Sales')
axes[0].grid(axis='y', linestyle='--')

# Scatter plot (second subplot)
sns.scatterplot(x='Height', y='Weight', size='Size', data=data_scatter, sizes=(20, 200), ax=axes[1])
axes[1].set_title('Height vs. Weight')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
axes[1].grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
