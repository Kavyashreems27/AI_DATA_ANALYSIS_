```python
# Ques 14 - Customized Line Plot using Matplotlib

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create a line plot with custom styles
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='orange')

# Adding titles and labels
plt.title('Customized Line Plot', fontsize=14, fontweight='bold')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Adding grid
plt.grid(True, linestyle=':', color='gray')

# Show plot
plt.show()
```
