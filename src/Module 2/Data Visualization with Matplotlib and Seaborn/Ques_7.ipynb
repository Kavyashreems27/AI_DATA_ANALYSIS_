```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with your actual height and weight data)
height = [165, 170, 155, 180, 160]
weight = [60, 75, 50, 85, 55]
size_data = [10, 15, 13, 18, 12]

# Create a Pandas DataFrame for easier handling with Seaborn
data = pd.DataFrame({'Height': height, 'Weight': weight, 'Size': size_data})

# Create the scatter plot with size variation
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Height', y='Weight', size='Size', data=data, sizes=(20, 200)) # Adjust sizes range as needed

# Customize the plot
plt.title('Height vs. Weight with Size Variation')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
