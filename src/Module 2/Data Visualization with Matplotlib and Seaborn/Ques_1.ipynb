```python
# Import the necessary library
import matplotlib.pyplot as plt

# Define the data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create the bar plot
plt.figure(figsize=(8, 6))
plt.bar(products, sales, color='skyblue')

# Add title and labels
plt.title('Sales of Products (in Thousands)', fontsize=16)
plt.xlabel('Products', fontsize=14)
plt.ylabel('Sales (Thousands)', fontsize=14)

# Show the plot
plt.show()
```
