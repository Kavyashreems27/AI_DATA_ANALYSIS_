```python
# Ques 16 - Combining Multiple Plots with Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.exp(-x / 3) * np.sin(x)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1
axs[0, 0].plot(x, y1, color='blue', linestyle='-', label='sin(x)')
axs[0, 0].set_title('Sine Function')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2
axs[0, 1].plot(x, y2, color='green', linestyle='--', label='cos(x)')
axs[0, 1].set_title('Cosine Function')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot 3
axs[1, 0].plot(x, y3, color='red', linestyle='-.', label='sin(2x)')
axs[1, 0].set_title('Double Frequency Sine')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot 4
axs[1, 1].plot(x, y4, color='purple', linestyle=':', label='Damped Sine')
axs[1, 1].set_title('Damped Sine Wave')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Main title and layout adjustment
fig.suptitle('Multiple Plots in One Figure', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
