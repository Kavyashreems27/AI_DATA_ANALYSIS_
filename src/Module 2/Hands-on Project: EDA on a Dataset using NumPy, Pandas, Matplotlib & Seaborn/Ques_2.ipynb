```python
# Ques 2 - Basic NumPy Operations on Iris Dataset

import numpy as np
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Convert numeric columns to NumPy array
numeric_data = iris.select_dtypes(include=[np.number]).to_numpy()

# Calculate basic statistics
mean_values = np.mean(numeric_data, axis=0)
median_values = np.median(numeric_data, axis=0)
std_values = np.std(numeric_data, axis=0)

# Display results
print("Mean of each numeric column:")
print(mean_values)

print("\nMedian of each numeric column:")
print(median_values)

print("\nStandard Deviation of each numeric column:")
print(std_values)
```
