```python
# Ques 8 - Advanced Data Visualization: Pair Plot

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Create a pair plot with species as hue
sns.pairplot(df, hue='species', palette='Set2', corner=True)

# Show the plot
plt.suptitle('Pairwise Relationships in Iris Dataset', fontsize=14, fontweight='bold', y=1.02)
plt.show()
```
