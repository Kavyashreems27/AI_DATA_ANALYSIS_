```python
# Ques 5 - Data Cleaning using Pandas: Handle Missing Values

import pandas as pd
import seaborn as sns

# Load the Iris dataset
df = sns.load_dataset('iris')

# Simulate missing data for demonstration (can be removed if your data already has NaNs)
df.loc[0:3, 'sepal_length'] = None
df.loc[10, 'petal_width'] = None

# Check for missing values
print("Missing values before imputation:")
print(df.isnull().sum())

# Impute missing numeric values with the median
df.fillna(df.median(numeric_only=True), inplace=True)

# Verify missing values are handled
print("\nMissing values after imputation:")
print(df.isnull().sum())

# Show the updated dataset (optional)
print("\nFirst 5 rows after cleaning:")
print(df.head())
```
