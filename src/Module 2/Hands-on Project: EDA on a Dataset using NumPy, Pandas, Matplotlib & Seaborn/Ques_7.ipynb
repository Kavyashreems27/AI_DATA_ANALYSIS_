```python
# Ques 7 - Group and Aggregate Data by Species

import pandas as pd
import seaborn as sns

# Load the Iris dataset
df = sns.load_dataset('iris')

# Group by 'species' and compute the mean of each numeric column
grouped_means = df.groupby('species').mean(numeric_only=True)

# Display the result
print("Average of each feature grouped by species:")
print(grouped_means)
```
