```python
# Ques 9 - Feature Engineering for Classification

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Create a new feature: petal_area = petal_length * petal_width
df['petal_area'] = df['petal_length'] * df['petal_width']

# Display first few rows to verify the new feature
print("First 5 rows with the new 'petal_area' feature:")
print(df[['petal_length', 'petal_width', 'petal_area', 'species']].head())

# Visualize the utility of petal_area using a boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal_area', data=df, palette='Set3')

# Add title and labels
plt.title('Petal Area by Species', fontsize=14, fontweight='bold')
plt.xlabel('Species')
plt.ylabel('Petal Area')

# Show the plot
plt.show()
```
