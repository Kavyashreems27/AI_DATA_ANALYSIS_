```python
# Ques 4 - Plot the Distribution of a Feature using Seaborn

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Plot the distribution of petal_length
plt.figure(figsize=(8, 5))
sns.histplot(df['petal_length'], kde=True, color='skyblue', bins=20)

# Add title and labels
plt.title('Distribution of Petal Length', fontsize=14, fontweight='bold')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
