```python
# Install necessary packages (run this cell once)
!pip install evidently pandas numpy --quiet

# Import required libraries
import pandas as pd
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset, DataQualityPreset

# Load datasets
# Adjust the path as needed based on your project structure
reference_data = pd.read_csv('../../data/reference.csv')  # training or baseline data
current_data = pd.read_csv('../../data/current.csv')      # new incoming data

# Create a report with both data drift and data quality checks
report = Report(metrics=[
    DataDriftPreset(),
    DataQualityPreset()
])

# Run the report
report.run(reference_data=reference_data, current_data=current_data)

# Save the report as an HTML file
report.save_html('data_drift_and_quality_report.html')

# (Optional) Display report inside notebook
report.show(mode='inline')
```
