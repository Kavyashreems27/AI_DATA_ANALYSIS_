```python
# Remove Duplicates & Fix Data Types in Pandas

import pandas as pd
import numpy as np

# Sample DataFrame with duplicates and incorrect data types
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve'],
    'Age': ['25', '30', '35', '25', '40'],
    'Gender': ['F', 'M', 'M', 'F', 'F'],
    'Score': ['85', '90', '95', '85', '100']
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# 1. Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicate rows:")
print(df_no_duplicates)

# 2. Fix data types
df_no_duplicates['Age'] = pd.to_numeric(df_no_duplicates['Age'], errors='coerce')
df_no_duplicates['Score'] = pd.to_numeric(df_no_duplicates['Score'], errors='coerce')

print("\nDataFrame after fixing data types:")
print(df_no_duplicates)

# 3. Check data types
print("\nData types of each column:")
print(df_no_duplicates.dtypes)
```
