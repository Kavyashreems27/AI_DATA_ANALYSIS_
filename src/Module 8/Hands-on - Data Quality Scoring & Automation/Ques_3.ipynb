```python
# Step 1: Install Great Expectations (uncomment if not already installed)
# !pip install great_expectations

# Step 2: Import Libraries
import pandas as pd
import great_expectations as ge
from great_expectations.dataset import PandasDataset

# Step 3: Load Dataset (replace with your actual dataset path)
data = {
    'id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'age': [25, 30, 35, 28, None],
    'email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'david@example.com', 'bob@example.com']
}
df = pd.DataFrame(data)

# Step 4: Convert to Great Expectations Dataset
ge_df = ge.from_pandas(df)

# Step 5: Define Expectations
ge_df.expect_column_to_exist('id')
ge_df.expect_column_to_exist('name')
ge_df.expect_column_to_exist('age')
ge_df.expect_column_to_exist('email')

ge_df.expect_column_values_to_not_be_null('id')
ge_df.expect_column_values_to_not_be_null('name')
ge_df.expect_column_values_to_not_be_null('age')

ge_df.expect_column_values_to_be_between('age', min_value=0, max_value=120)
ge_df.expect_column_values_to_match_regex('email', r'^[\w\.-]+@[\w\.-]+\.\w+$')
ge_df.expect_column_values_to_be_unique('id')

# Step 6: Run Validation
validation_result = ge_df.validate()

# Step 7: Display Results
import json
print(json.dumps(validation_result, indent=2))
```
