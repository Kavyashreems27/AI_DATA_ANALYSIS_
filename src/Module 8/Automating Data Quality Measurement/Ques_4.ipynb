```python
# Ques_4.ipynb
# Automating Data Quality Measurement with Python

import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('data.csv')  # Replace with your file name/path

# Display basic info
print("### Data Preview ###")
print(df.head())

# Functions to compute data quality metrics
def calculate_completeness(column):
    return column.notnull().sum() / len(column)

def calculate_uniqueness(column):
    return column.nunique() / len(column)

def calculate_consistency(column):
    if column.dtype == 'object':
        return column.str.strip().str.lower().value_counts(normalize=True).max()
    return np.nan

def calculate_validity(column):
    # Placeholder: define actual rules if required
    return np.nan

# Automate calculation for each column
metrics = []
for col in df.columns:
    completeness = calculate_completeness(df[col])
    uniqueness = calculate_uniqueness(df[col])
    consistency = calculate_consistency(df[col])
    validity = calculate_validity(df[col])

    available_metrics = [v for v in [completeness, uniqueness, consistency, validity] if not pd.isna(v)]
    dqi = sum(available_metrics) / len(available_metrics) if available_metrics else np.nan

    metrics.append({
        'Column': col,
        'Completeness': round(completeness, 2),
        'Uniqueness': round(uniqueness, 2),
        'Consistency': round(consistency, 2) if not pd.isna(consistency) else None,
        'Validity': round(validity, 2) if not pd.isna(validity) else None,
        'DQI': round(dqi, 2)
    })

# Create DataFrame for metrics
dqi_df = pd.DataFrame(metrics)

# Display results
print("\n### Data Quality Report ###")
print(dqi_df)

# Optionally save the report
dqi_df.to_csv("data_quality_report.csv", index=False)
print("\nReport saved to 'data_quality_report.csv'")
```
