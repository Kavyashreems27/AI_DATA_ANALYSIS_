```python
# Ques_3.ipynb
# AI-Based Anomaly Detection using Isolation Forest

import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file
df = pd.read_csv('data.csv')  # Replace with your actual file path

# Show basic info
print("### Data Preview ###")
print(df.head())

# Preprocess: Select numerical columns only
numerical_df = df.select_dtypes(include=[np.number]).dropna()

if numerical_df.empty:
    print("No numerical columns found or data is empty after dropping NaNs.")
else:
    # Fit Isolation Forest
    model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
    model.fit(numerical_df)

    # Predict anomalies: -1 = anomaly, 1 = normal
    df['Anomaly'] = model.predict(numerical_df)
    df['Anomaly'] = df['Anomaly'].map({1: 'Normal', -1: 'Anomaly'})

    # Show anomaly results
    print("\n### Anomaly Summary ###")
    print(df['Anomaly'].value_counts())

    # Show anomalous records
    print("\n### Sample Anomalies ###")
    print(df[df['Anomaly'] == 'Anomaly'].head())

    # Optional Visualization (2D plot with first two numeric features)
    if numerical_df.shape[1] >= 2:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(
            x=numerical_df.iloc[:, 0],
            y=numerical_df.iloc[:, 1],
            hue=df['Anomaly'],
            palette={'Normal': 'green', 'Anomaly': 'red'}
        )
        plt.title("Anomaly Detection using Isolation Forest")
        plt.xlabel(numerical_df.columns[0])
        plt.ylabel(numerical_df.columns[1])
        plt.legend()
        plt.show()
```
