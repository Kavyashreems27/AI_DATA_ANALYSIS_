```python
# Ques_2.ipynb
# Custom Expectation with Great Expectations

import pandas as pd
import great_expectations as ge
from great_expectations.dataset import PandasDataset

# Load data
df = pd.read_csv('data.csv')  # Replace with your CSV file
print("### Data Preview ###")
print(df.head())

# Define a custom expectation by extending PandasDataset
class CustomDataset(PandasDataset):

    def expect_column_values_to_be_uppercase(self, column):
        """Expect that all values in a column are uppercase (strings only)."""
        def is_upper(val):
            if isinstance(val, str):
                return val.isupper()
            return False

        return self.expect_column_values_to_match_function(
            column, is_upper,
            result_format="SUMMARY"
        )

# Convert to CustomDataset
custom_df = CustomDataset(df)

# Run the custom expectation (example: validate "NAME" column is all uppercase)
result = custom_df.expect_column_values_to_be_uppercase('NAME')  # Replace 'NAME' with your actual column
print("\n### Custom Expectation Result ###")
print(result)

# Optional: View summary
if result['success']:
    print("\nAll values passed the expectation.")
else:
    print("\nSome values failed the expectation.")
    print("Unexpected values:", result['result']['unexpected_list'])
```
