```python
# Ques_1.ipynb
# Basic Data Profiling of a CSV File

import pandas as pd

# Upload CSV file
file_path = 'data.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# 1. Basic Information
print("### Basic Info ###")
print(f"Shape: {df.shape}")
print("\nData Types:")
print(df.dtypes)
print("\nNumber of Duplicated Rows:", df.duplicated().sum())

# 2. Missing Values
print("\n### Missing Values ###")
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing, 'Missing %': missing_percent})
print(missing_df)

# 3. Unique Values
print("\n### Unique Values per Column ###")
print(df.nunique())

# 4. Descriptive Statistics
print("\n### Descriptive Statistics ###")
print(df.describe(include='all').transpose())

# 5. Correlation Matrix (for numerical columns)
print("\n### Correlation Matrix ###")
print(df.corr(numeric_only=True))

# 6. Top 5 Frequent Values for Each Column
print("\n### Top 5 Frequent Values per Column ###")
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts(dropna=False).head(5))
```
