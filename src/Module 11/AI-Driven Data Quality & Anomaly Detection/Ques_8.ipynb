```python
# Removing Stopwords

# Step 1: Import Required Libraries
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Step 2: Sample Text Data
data = {
    'text': [
        "This is a simple sentence with some stopwords.",
        "Stopwords are words which do not contain important meaning.",
        "Removing them can help in better analysis.",
        "Text preprocessing is essential in NLP tasks.",
        "We will remove stopwords from this sentence."
    ]
}
df = pd.DataFrame(data)

# Step 3: Define Function to Remove Stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text.lower())  # Tokenize and lowercase
    filtered_text = [word for word in word_tokens if word.isalpha() and word not in stop_words]
    return ' '.join(filtered_text)

# Step 4: Apply Function to DataFrame
df['cleaned_text'] = df['text'].apply(remove_stopwords)

# Step 5: Show Results
print("Original vs. Cleaned Text:")
for original, cleaned in zip(df['text'], df['cleaned_text']):
    print(f"\nOriginal: {original}")
    print(f"Cleaned:  {cleaned}")
```
