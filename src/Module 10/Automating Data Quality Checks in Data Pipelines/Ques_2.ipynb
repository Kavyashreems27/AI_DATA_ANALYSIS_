```python
# Ques_2.ipynb

# Step 1: Install Great Expectations
!pip install great_expectations

# Step 2: Import libraries
import great_expectations as ge
import pandas as pd
from great_expectations.dataset import PandasDataset

# Step 3: Create or load a sample dataset
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Edward"],
    "age": [25, 30, 35, None, 40],
    "email": ["alice@example.com", "bob@example.com", "charlie@example.com", None, "edward@example.com"]
}
df = pd.DataFrame(data)

# Step 4: Convert DataFrame into Great Expectations dataset
ge_df = ge.from_pandas(df)

# Step 5: Define expectations
ge_df.expect_column_values_to_not_be_null("name")
ge_df.expect_column_values_to_be_between("age", min_value=18, max_value=60)
ge_df.expect_column_values_to_match_regex("email", r"[^@]+@[^@]+\.[^@]+")

# Step 6: Validate the dataset and show results
results = ge_df.validate()
results
```
