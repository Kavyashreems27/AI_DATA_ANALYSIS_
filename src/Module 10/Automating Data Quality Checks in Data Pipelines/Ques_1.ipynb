```python
from airflow import DAG
from airflow.utils.dates import days_ago
from airflow.providers.great_expectations.operators.great_expectations import GreatExpectationsOperator
from datetime import timedelta

default_args = {
    'owner': 'airflow',
    'start_date': days_ago(1),
    'retries': 1,
    'retry_delay': timedelta(minutes=5),
}

with DAG(
    dag_id='great_expectations_data_quality_dag',
    default_args=default_args,
    schedule_interval=None,
    catchup=False,
    description='Run data quality check using Great Expectations and Apache Airflow',
) as dag:

    gx_task = GreatExpectationsOperator(
        task_id='run_data_quality_check',
        checkpoint_name='my_checkpoint',
        data_context_root_dir='/usr/local/airflow/great_expectations',
        fail_task_on_validation_failure=True
    )

    gx_task
```
