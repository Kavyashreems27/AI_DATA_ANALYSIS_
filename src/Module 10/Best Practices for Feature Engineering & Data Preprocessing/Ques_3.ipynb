```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
from sklearn.datasets import load_iris

# Load sample data (you can replace this with your own CSV/data source)
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)

# Display original feature shape
print("Original shape:", X.shape)

# Apply Variance Threshold
# Threshold = 0 means remove features with the same value in all rows
selector = VarianceThreshold(threshold=0.2)  # You can adjust this threshold
X_reduced = selector.fit_transform(X)

# Get selected feature names
selected_features = X.columns[selector.get_support(indices=True)]
X_selected = pd.DataFrame(X_reduced, columns=selected_features)

# Display reduced feature shape and the retained features
print("Reduced shape:", X_selected.shape)
print("Selected features:")
print(selected_features)

# Optional: Display final dataset
print("\nFinal dataset after removing low-variance features:")
print(X_selected.head())
```
